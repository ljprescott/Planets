1. 	What is a Graphical User Interface (GUI) and how does that differ from the Command Line Interface (CLI)?

	A GUI (pronounced gooey) is a program with drop down menus that guides the user with choices. The CLI, on the other hand, is a program that accepts commands from a use in the form of text. The CLI does not have drop down options like a GUI. A GUI is a bit more restrictive than the CLI but typically is more user friendly.

19.	What is the difference between ```>``` and ```>>```?

	> and >> are both ways of directing an output into a file (either by creating a new file or modifying an existing file)
	a single > will overwrite the existing file (if it exists) and a double >> will append the output to the end of the file (if it exists)

26.	What are the benefits of writing and running a script over typing the code in?

27.	When looking at a script what does ```#``` mean? Why would you use one?

	A hash will denote a nonexecutable statement, it's typically used to 'hash' out text as a comment so the computer doesn't try to run the text as code but can also be used to break up code visually or can be used in markdown to change header sizes

28.	What does ```grep``` stand for and what does it do?

	grep (pronounced "grr ehp", not "jee rep" like I first assumed) stands for the commands g, re, and p all together: global search, regular expression, print
	Basically, it searches through a file for a string of text (e.g. 'ATGA') and returns any matches per the regular expression

29.	What does ```find``` do?

	find will search through specified directories (typically all but can be limited) for  a file based on a regular expression e.g. 'file.txt' vs 'file*"
